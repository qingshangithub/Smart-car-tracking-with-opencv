```python
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.cross_validation import train_test_split
from sklearn.metrics import r2_score
import numpy as np
import pandas as pd
import pickle
import itertools
```

```python
x = pickle.load(open('x.con','rb'))
y = pickle.load(open('y.con', 'rb'))
```

```python
x_array = np.array(x)
y_array = np.array(y)
x_train, x_test, y_train, y_test = train_test_split(x_array, y_array, test_size=0.1)
Y_random = np.random.uniform(-1,1,len(y_test))
print(x_train.shape, x_test.shape)
```

```python
rf = RandomForestRegressor(n_jobs=-1, n_estimators=25, max_depth=20)
rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)
print(r2_score(y_pred=Y_random, y_true=y_test))
print(r2_score(y_pred=y_pred, y_true=y_test))
```

```python
mlp = MLPRegressor(hidden_layer_sizes=(10,), algorithm='l-bfgs'
                  ,alpha=2)
mlp.fit(x_train, y_train)
y_pred = mlp.predict(x_test)
print(r2_score(y_pred=Y_random, y_true=y_test))
print(r2_score(y_pred=y_pred, y_true=y_test))
```

```python
hid_one = range(10,20)
alpha_range = np.linspace(1.5,3, 20)
mlp_search_result = []
for h1, a in itertools.product(hid_one,alpha_range):
    r2 = 0
    if r2 <= 0.1:
        mlp = MLPRegressor(hidden_layer_sizes=(h1,), algorithm='l-bfgs',alpha=a)
        mlp.fit(x_train, y_train)
        y_pred = mlp.predict(x_test)
        r2 = r2_score(y_pred=y_pred, y_true=y_test)
    print(r2)
    mlp_search_result.append((h1, a, r2))
```
